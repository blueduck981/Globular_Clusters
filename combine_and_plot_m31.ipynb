```python
#This code will collect abundance results, organize them, and make figures.
# The results have originally been stored by individual Cluster, this code will put them
# into a large table and reindex by Element in order to make individual
# Element figures
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt
import glob
import pandas as pd
plt.rcParams['font.family']='sans-serif'  # this changes all fonts
import scipy
from scipy import stats
```

```python
#Load in the Files with [Fe/H] information
m31tab=pd.read_table('m31_fe_new.txt',skipinitialspace=1,sep=' ')
mwtab=pd.read_table('mw_all_fe.txt',skipinitialspace=1,sep=' ')

#M31 cluster and [Fe/H] information
m31tab
```

```python
#isolate the name column, calculate the total error
knames=m31tab['Name']
first=m31tab["e1"]/np.sqrt(m31tab["num"]-1)
m31tab['Ter']=np.sqrt(first*first+m31tab['eage']*m31tab['eage'])
mnames=mwtab['cl']
mfirst=mwtab['Er']/np.sqrt(mwtab['N']-1)
mwtab['FTer']=np.sqrt(mfirst*mfirst+mwtab['SigA']*mwtab['SigA'])
```

```python
#find individual cluster abundance files, should be in same order as in m31tab
files=glob.glob("*.abund_diff")
mfiles=glob.glob("*diffab.txt") 
```

```python
#set names of columns in abund files
unames=['Element', 'Ion', 'logep', 'sigma_age','sigma_mean','total_error'
        ,'n_lines','xfei','limit_flag' ]
```

```python
#start a list to hold the dataframes of abundance data files
datalist=[]
mdatalist=[]
```

```python
#run through the abundance file list, read the information and 
#add a column with the name of the cluster so that we can merge later

for i,item in enumerate(files):
    abundtab=pd.read_table(item,sep=' ', skipinitialspace=1,names=unames,skiprows=1)
    abundtab['Name']=knames[i]
    abundtab[(abundtab['Element']=='Ti') & (abundtab['Ion']=='II')]
    #Need to distinguish between Ti I and Ti II by element, so changing Ti II
    abundtab.loc[(abundtab.Element=='Ti') & (abundtab.Ion=='II'),'Element']='Ti2'
    datalist.append(abundtab)

#Example Cluster abundance File
abundtab
```

```python
#Second, run through Milky Way files
for i,item in enumerate(mfiles):
    mabundtab=pd.read_table(item,sep=' ', skipinitialspace=1,names=unames,skiprows=1)
    mabundtab['cl']=mnames[i]
    mabundtab.loc[(mabundtab.Element=='TiII') & (mabundtab.Ion=='II'),'Element']='Ti2'
    mdatalist.append(mabundtab)
```

```python
#concatenate abundance files of all GCs into one table (MW)
mbigframe=pd.concat(mdatalist)
```

```python
#Merge the abundance file information with the [Fe/H] and properties table
#Merge on name of cluster (MW)
mfabund=pd.merge(mwtab,mbigframe,on='cl',how='outer')
```

```python
#concatenate abundance files of all GCs into one table (M31)
bigframe=pd.concat(datalist)
```

```python
#Merge the abundance file information with the [Fe/H] and properties table
#Merge on name of cluster (M31)
fullabundtab=pd.merge(m31tab,bigframe, on='Name',how='outer')
fullabundtab
```

```python
#array to loop through individual elements 
el_array=["Ca","Si","Ba","Ni","Cr","Sc","Mn","Mg","Al","Na","V","Zr",
          "Co","Ti","Ti2","Y","Nd","Cu","La","Eu"]
```

```python
#Some analysis to check on data
#2 Sample t-test example to check if MW and M31 results are consistent
stats.ttest_ind(fullabundtab[fullabundtab.Element=='Sc'].xfei,
                mfabund[mfabund.Element=='Sc'].xfei,equal_var=False,nan_policy='omit')
```

```python
#Run the t-test for every element and store results
tvalues=[]
pvalues=[]
for el in el_array:
    t,p=stats.ttest_ind(fullabundtab[fullabundtab.Element==el].xfei,
                mfabund[mfabund.Element==el].xfei,equal_var=False,nan_policy='omit')
    tvalues.append(t)
    pvalues.append(p)
statsdf=pd.DataFrame({'Element':el_array,'tvalues':tvalues, 'pvalues':pvalues})
statsdf=statsdf[['Element','tvalues','pvalues']]
statsdf
```

```python
#calculating the mean of Ba but eliminating outliers
cut=fullabundtab[(fullabundtab.Element=='Ba') & (fullabundtab.xfei < 0.65)].xfei
c, low, up=stats.sigmaclip(cut,3.,3.)
c.mean()
```

```python
#reindex by element
bigtable=fullabundtab.set_index('Element')
mwtable=mfabund.set_index('Element')
bigtable
```

```python
#Save table to file
bigtable.to_csv('test_bigtable.csv',na_rep='NaN')
```

```python
#Loop through elements, create one Figure for each element and plot both 
#MW and M31.  Check for measurements that are upper limits to plot them with downward arrows
#Use a different color for the GCs above [Fe/H]=-0.30 (B193)
#Use a different color for G002 to call attention to it
for i,el_name in enumerate(el_array):
    #Make the Ti Name prettier
    el_title='Ti II' if el_name=='Ti2' else el_name
    #Figure set up
    fig1=plt.figure(1,facecolor='white',edgecolor='black')
    plt.gcf().subplots_adjust(bottom=0.15)
    plt.axis([-2.5,0.5,-1.,1.])
    plt.xlabel(r'[Fe/H]',fontsize=18)
    plt.ylabel(r'['+el_title+r'/Fe] ',fontsize=18)
    plt.plot([-10,10],[0,0],'--',c='black')
    plt.plot([0,0],[-10,10],'--',c='black')

    #Plot MW 
    plt.errorbar(mwtable.ix[el_name,'Fe'],
                 mwtable.ix[el_name,'xfei'],
                 marker='o',linestyle='None',xerr=mwtable.ix[el_name,'FTer'],
                 yerr=mwtable.ix[el_name,'total_error'],
                 color='#cccccc',ecolor='black',ms=8)   

    #Plot M31
    plt.errorbar(bigtable[bigtable.limit_flag !='LIM'].ix[el_name,'[Fe/H]'],
                 bigtable[bigtable.limit_flag !='LIM'].ix[el_name,'xfei'],
                 marker='s',linestyle='None',
                 xerr=bigtable[bigtable.limit_flag !='LIM'].ix[el_name,'Ter'],
                 yerr=bigtable[bigtable.limit_flag !='LIM'].ix[el_name,'total_error'],
                 color='#3366ff',ecolor='black',ms=8)   
   
    #Plot upper limits in M31
    if any(bigtable.ix[el_name].limit_flag =='LIM'):
             plt.errorbar(bigtable[bigtable.limit_flag =='LIM'].ix[el_name,'[Fe/H]'],
                     bigtable[bigtable.limit_flag =='LIM'].ix[el_name,'xfei'],
                     marker='s',linestyle='None',
                     xerr=bigtable[bigtable.limit_flag =='LIM'].ix[el_name,'Ter'],
                     yerr=0.15, uplims=True, color='#3366ff',linewidth=2,
                          ecolor='black',ms=8,capsize=5)   
    
    #Call out G002
    if any(bigtable[bigtable.Name=='G002'].index==el_name):
        plt.plot(bigtable[bigtable.Name=='G002'].ix[el_name,'[Fe/H]'],
                 bigtable[bigtable.Name=='G002'].ix[el_name,'xfei'],
                 color='#ffff00', ms=9,marker="s",linestyle="None",markeredgewidth=1,
                 zorder=3)
    #Call out B193
    if any(bigtable[bigtable.Name=='B193-G244'].index==el_name):
        plt.plot(bigtable[bigtable.Name=='B193-G244'].ix[el_name,'[Fe/H]'],
                 bigtable[bigtable.Name=='B193-G244'].ix[el_name,'xfei'],
                 color='#804000', ms=8,marker="s",linestyle="None",markeredgewidth=1,
                 zorder=3)    
    
    plt.show()
    #plt.savefig('Abundance_M31_'+el_name+'.ps',orientation='portrait',
    #     format='ps',edgecolor='black')
    #plt.close()
    
```
